```python
import pandas as pd
import mysql.connector
import json

HOST = ''
USERNAME = ''
PASSWORD = ''
DB = ''
```

```python
#connect to the DB
mydb = mysql.connector.connect(
            host=HOST,
            user=USERNAME,
            passwd=PASSWORD,
            database=DB
            )

if mydb.is_connected():
    try:
        print('Connected to MySQL database')
        mycursor = mydb.cursor()
        mycursor.execute("""select * from department_data;""")

        #Result will have elements as tuples
        result = mycursor.fetchall()
        df = pd.DataFrame(result,columns=['dept_id','dept_name','dept_head'])
        df.to_csv('department_data.csv')
    except:
        print("Unable to Connect to MYSQL DB")
    finally:
            mycursor.close()
            mydb.close()
        
df.head()
```
